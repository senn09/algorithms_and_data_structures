#### Starter Code for Question 1

```python
class TreeNode(object):
    def __init__(self, val = 0, left = None, right = None):
        self.val = val
        self.left = left
        self.right = right
```

```python
def is_duplicate(root: TreeNode) -> int:
  checked_values = set()

  def traverse(node, bool_is_duplicate: bool) -> bool:
    if node: # Check if node is empty
        if node.val in checked_values:
            print(f"Duplicate found for value: {node.val}")
            return True
        else:
          checked_values.add(node.val)
          # Needs only one True to return True using boolean OR operation
          return traverse(node.left, bool_is_duplicate) or traverse(node.right, bool_is_duplicate)
    else:
        return False

  return traverse(root, False)
```

```python
# Test: False

root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.left = TreeNode(6)

is_duplicate(root)
```

```python
# Test: True

root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(2)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.left = TreeNode(6)

is_duplicate(root)
```

### Copied Solution

```python
from typing import List

def move_zeros_to_end(nums: List[int]) -> List[int]:
    # TODO
    insert_pos = 0

    # Step 1: move non-zeros forward
    for num in nums:
        if num != 0:
            nums[insert_pos] = num
            insert_pos += 1

    # Step 2: fill remaining with zeros
    for i in range(insert_pos, len(nums)):
        nums[i] = 0

    return nums
```


-   Explain why their solution works in your own words.


### Solution Explanation

```python
'''
Step 1 handles moving the non-zero elements to the left side of the list while maintaining their order.
- num != 0 ensures that only non-zero elements are moved
- the loop itself maintains the order of the non-zero elements
- while the insert_pos maintains the index of where the next non-zero element should be placed


Step 2 handles filling the remaining positions in the list with zeros.
- insert_pos only increments due to non-zero elements
- the range(insert_pos, len(nums)) will have the correct starting index and length (total number of zeros)
'''
```


-   Explain the problem’s time and space complexity in your own words.


### Time and Space Complexity

```python
'''
Time complexity: O(n) 
- we iterate through the list twice  -> O(n + n) = O(n)
- operations inside the loops are constant time and do not affect the overall time complexity.

Space complexity: O(1)
- The list is modified in place, so no additional space is used.
'''
```
